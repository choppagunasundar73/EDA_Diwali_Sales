```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('/Users/choppagunasundar73/Downloads/Diwali Sales Data.csv', encoding='ISO-8859-1')
df.shape
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
# From above graph the purchasing power of females greater than that of males
```

```python
gx = sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in gx.containers:
    gx.bar_label(bars)
```

```python
df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
#female buyers from the age group of 26-35 spend the most
```

```python
sales_by_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(12)
sns.set(rc={'figure.figsize':(19,5)})
sns.barplot(data=sales_by_state,x='State',y='Orders')
```

```python
#From above graphs we can see that Uttar Pradesh,Maharashtra,Karnataka spend the most for diwali
```

```python
mx = sns.countplot(data = df,x='Marital_Status')
sns.set(rc={'figure.figsize':(7,3)})
for bars in mx.containers:
    mx.bar_label(bars)
```

```python
sales_marital=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,3)})
sns.barplot(data=sales_marital,x='Marital_Status',y='Amount',hue='Gender')
```

```python
sales_by_state = df.groupby(['Occupation'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(12)
sns.set(rc={'figure.figsize':(19,5)})
sns.barplot(data=sales_by_state,x='Occupation',y='Orders')
```

```python
##People in IT sector,Healthcare,Aviation seem to place more orders
```

```python
plt.figure(figsize=(22, 6))
mx = sns.countplot(data=df, y='Product_Category')

for bar in mx.patches:
    yval = bar.get_height()
    mx.annotate(yval, xy=(bar.get_width(), bar.get_y() + bar.get_height() / 2),
                xytext=(5, 0), textcoords='offset points', va='center', ha='left')

plt.show()
```

```python
sales_by_state = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_by_state,x='Product_Category',y='Amount')
```

```python
##It can be seen that people sre spending more on Food, CLothing and Electronics the most
```

```python
sales_by_state = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_by_state,x='Product_ID',y='Orders')
```

```python
##Conclusion : Married Women of Age-Group 26-35 from UP,Maharashtra,Karnataka working in IT,Healthcare,Avaiation sectors tend to purchase more during Diwali from the Food,Clothing and Electronics categories
```
